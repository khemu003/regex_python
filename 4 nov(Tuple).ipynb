### how can we create a user a defined Tuple?

```python
a = []
size = int(input("Enter tuple size : "))
for i in range(size):
    val = input("Enter item : ")
    a.append(val)
b = tuple(a)
print(b)
```

```python
a = ('sam', '34', 'rahul', '56')
b = list(a)
c = []
d = []
print(b)
for i in range(len(b)):
    if (b[i].isdigit()==True):
        c.append(b[i])
    elif (b[i].isalnum()==True):
        d.append(b[i])

e = tuple(c)
f = tuple(d)
print(e)
print(f)
```
